# Forecasting Sales Superstore

```python
df_date.rename(columns = {'Order Date':'ds','Sales':'y'}, inplace = True)
```

```python
df_date
```

```python
from prophet import Prophet
```

```python
pre = Prophet()
```

```python
pre.fit(df_date)
future = pre.make_future_dataframe(periods=36)

forecast = pre.predict(future)

pre.plot(forecast)
plt.legend(['Actual','Predicted'])
plt.title("Prophet Model Forecasts on Sales Superstore")
```
